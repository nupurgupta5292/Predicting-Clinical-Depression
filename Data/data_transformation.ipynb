```python
# Importing dependencies

import pandas as pd
```

```python
# Reading both csv datasets
test_df = pd.read_csv("test.csv")
train_df = pd.read_csv("train.csv")
test_df
```

```python
train_df
```

```python
# Combining Datasets
combo = [test_df, train_df]
depression_df = pd.concat(combo)
depression_df
```

```python
# Dropping duplicates, if any
depression_df = depression_df.drop_duplicates()
depression_df
```

```python
depression_df.columns
```

```python
# Dropping columns not relevant to our analysis and reducing the number of input parameters
depression_df = depression_df.drop(['surveyid', 'village', 'survey_date', 'hh_totalmembers','early_survey','day_of_week',
                                   'asset_livestock','asset_niceroof','cons_ownfood','ent_wagelabor','ent_ownfarm',
                                   'ent_business','ent_nonagbusiness', 'ent_employees','ent_farmrevenue', 'ent_farmexpenses',
                                    'ent_animalstockrev','fs_adskipm_often', 'fs_chskipm_often','med_expenses_hh_ep',
                                    'med_expenses_sp_ep', 'med_expenses_child_ep','med_portion_sickinjured',
                                    'med_port_sick_child', 'med_afford_port','med_healthconsult','med_vacc_newborns',
                                    'med_child_check','ed_expenses_perkid', 'ed_sch_missedpc','labor_primary',
                                    'ed_work_act_pc', 'wage_expenditures','given_mpesa','amount_given_mpesa',
                                    'received_mpesa','amount_received_mpesa', 'net_mpesa','saved_mpesa','amount_saved_mpesa'],
                                    axis=1)
depression_df
```

```python
# Renaming some columns for better understanding
depression_df = depression_df.rename(columns={'femaleres':'sex','married':'marital_status',
                             'hhsize':'household_size','edu':'years_of_edu'})
depression_df
```

```python
# Dropping rows with no values for 'Depressed' column, as they won't be useful as an input to our model
depression_df = depression_df[depression_df.depressed.notnull()]
depression_df
```

```python
depression_df.columns
```

```python
# Checking for datatypes of each column
depression_df.dtypes
```

```python
# Correcting datatypes for easy analysis, wherever necessary
depression_df = depression_df.apply(pd.to_numeric, errors='coerce')
```

```python
depression_df.dtypes
```

```python
# Identifying columns with null values
depression_df.isnull().sum()
```

```python
# Filling Missing values for columns where average of the column values may be a better representation of the data
depression_df['cons_alcohol'].fillna(depression_df['cons_alcohol'].mean(), inplace=True)
depression_df['cons_tobacco'].fillna(depression_df['cons_tobacco'].mean(), inplace=True)
depression_df['fs_meat'].fillna(depression_df['fs_meat'].mean(), inplace=True)
```

```python
depression_df.isnull().sum()
```

```python
# Filling 0 for columns, where such assumption would make most sense
depression_df['med_u5_deaths'] = depression_df['med_u5_deaths'].fillna(0)
depression_df['fs_enoughtom'] = depression_df['fs_enoughtom'].fillna(0)
depression_df['fs_sleephun'] = depression_df['fs_sleephun'].fillna(0)
```

```python
depression_df.isnull().sum()
```

```python
# Filling number of children wise averages for columns, where such assumption would make most sense
depression_df['cons_med_children'] = depression_df['cons_med_children'].fillna(depression_df.groupby(depression_df['children'])['cons_med_children'].transform('mean'))
depression_df['cons_med_children'] = depression_df['cons_med_children'].fillna(0)

depression_df['fs_chwholed_often'] = depression_df['fs_chwholed_often'].fillna(depression_df.groupby(depression_df['children'])['fs_chwholed_often'].transform('mean'))
depression_df['fs_chwholed_often'] = depression_df['fs_chwholed_often'].fillna(0)

depression_df['ed_expenses'] = depression_df['ed_expenses'].fillna(depression_df.groupby(depression_df['hh_children'])['ed_expenses'].transform('mean'))
depression_df['ed_expenses'] = depression_df['ed_expenses'].fillna(0)

depression_df['ed_schoolattend'] = depression_df['ed_schoolattend'].fillna(depression_df.groupby(depression_df['hh_children'])['ed_schoolattend'].transform('mean'))
depression_df['ed_schoolattend'] = depression_df['ed_schoolattend'].fillna(0)
```

```python
depression_df.isnull().sum()
```

```python
# Filling number of household wise average for column, where such assumption would make most sense
depression_df['med_sickdays_hhave'] = depression_df['med_sickdays_hhave'].fillna(depression_df.groupby(depression_df['household_size'])['med_sickdays_hhave'].transform('mean'))
depression_df['med_sickdays_hhave'] = depression_df['med_sickdays_hhave'].fillna(0)
depression_df[['household_size','med_sickdays_hhave']]
```

```python
depression_df.isnull().sum()
```

```python
# Exporting final, cleaned dataframe to csv
depression_df.to_csv('depression_data_cleaned.csv',index=False)
```
