```python
# Importing dependencies

import pandas as pd
```

```python
# Reading both csvs
test_df = pd.read_csv("test.csv")
train_df = pd.read_csv("train.csv")
test_df
```

```python
train_df
```

```python
combo = [test_df, train_df]
depression_df = pd.concat(combo)
depression_df
```

```python
# Dropping duplicates, if any
depression_df = depression_df.drop_duplicates()
depression_df
```

```python
depression_df.columns
```

```python
# Dropping first two columns as they're IDs and not relevant to our analysis
depression_df = depression_df.drop(['surveyid', 'village', 'survey_date', 'hh_totalmembers','early_survey','day_of_week'], axis=1)
depression_df
```

```python
depression_df = depression_df.rename(columns={'femaleres':'sex','married':'marital_status',
                             'hhsize':'household_size','edu':'years_of_edu'})
depression_df
```

```python
depression_df = depression_df[depression_df.depressed.notnull()]
depression_df
```

```python
depression_df = depression_df.fillna(0)
```

```python
depression_df.dtypes
```

```python
depression_df.to_csv('depression_data_cleaned.csv',index=False)
```
